```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color palette
color_palette = {
    'Axillae': (127/255, 128/255, 0/255, 0.5),
    'Genital': (1, 0, 0, 0.5),
    'Mammary': (50/255, 1, 0, 0.5),
    'Buttocks': (28/255, 144/255, 1, 0.5)
}
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6), dpi = 300)

# Create a custom legend
for i, (label, color) in enumerate(color_palette.items()):
    ax.bar(0, 0, color=color, label=label, alpha=0.5)

# Customize the legend
ax.legend(title='Body Areas', loc='center', frameon=True, fontsize=12)

# Remove axis
ax.axis('off')


fig.savefig("t.png")
# Show the plot
plt.tight_layout()
plt.show()
```
